```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('Salary_Data.csv')
df
```

```python
X=df['YearsExperience']
y=df['Salary']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X,y)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
theta_0=0.0
theta_1=0.0
alpha=0.0001
```

```python
result=theta_0+(theta_1*X)
print("h(x):\n",result)
```

```python
error=result-y
print("Error vector:\n",error)
#res=np.linalg.norm(sum(error))
m=len(X)
cost_f=(1/(2*m)*sum(error**2))
print("j=",cost_f)
```

```python
d_theta_0=(1/m)*sum(error)
d_theta_1=(1/m)*sum((error)*X)
print("d_theta_0:",d_theta_0)
print("d_theta_1:",d_theta_1)
```

```python
theta_0_new=theta_0-(alpha*d_theta_0)
theta_1_new=theta_1-(alpha*d_theta_1)
print("theta_0_new:",theta_0_new)
print("theta_1_new:",theta_1_new)
```
